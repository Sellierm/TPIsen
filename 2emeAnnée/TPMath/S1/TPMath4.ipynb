$ \text{On observe la 2.pi periodicité de la fonction(pour les réeles). Pour les imaginaires, les modules tendent vers +l'infini. On a des zeros tout les pi/2 + k*pi }$

```python
f(z) = (z^2 + z + 1)/(2-z) 
complex_plot(f, (-2,2), (-2,2))
```

```python
var('n')
def fN(z,N):
    return 1/2 + 3/4*z +7/2*sum(z^n/2^n, n,2,N)

complex_plot(fN(z,100),(-5,5),(-5,5),aspect_ratio=1,plot_points=500)
```

```python
var('n')
def gN(z,N):
    return sum( ((-1)^n/3^(n+1))*(z-3)^n,n,0,N)
complex_plot(gN(z,30),(-2,8),(-6,6),plot_points=500,aspect_ratio=1)
```

$ \text{On constate que le rayon de convergence est de 3}$

```python
f(z) = arctan(z)
complex_plot(f, (-2,2), (-2,2))
```

```python
reset()

def F(X):
    return numerical_integral(sin(x)/x, 0, X)[0]

plot(sin(x)/x, (0, 50), color='blue') +plot(F, (0, 50), color='red')

```

On peut conjecturer que l'intégralle va converger vers π/2 

```python
X = 1
n = X/pi
RX=2/((n+1)* pi)
eps = 10^-3

while RX > eps:
    X+=1
    n=X/pi
    RX = 2/((n+1)* pi)
    
show("Pour n = ",n,", On a F(x) = ", numerical_approx(F(X), digits=4) )
```

<h2>
    a)
</h2>
<h5>
    $\quad\quad \displaystyle{I:=\int_{0}^{+\infty}\frac{1-e^{-x}}{1+x^{2}}dx, \quad\quad \lvert R(X) \rvert \leq \int_{X}^{+\infty}\frac{1}{(1+x^{2})}dx}=\frac{\pi}{2}-arctan(X)$.
</h5>

```python
reset()

X = 1
RX = pi / 2 - arctan(X)
eps = 10^-3

def f(x):
    return (1 - exp(-x)) / (1 + x^(2))

def F(X):
    return numerical_integral(f, 0, X)[0]



plot(f, (0, 25), color='blue') +plot(F, (0, 25), color='red')
```

On peut conjecturer que l'intégralle va converger vers 0.95

```python
while RX > eps:
    X+=1
    RX = pi / 2 - arctan(X)
    
show("Pour X=", X, ", on a I=", numerical_approx(F(X), digits=4), )
```

La réponse est cohérente avec nos observations


<h2>

<h5>
    $\quad\quad \displaystyle{I:=\int_{0}^{+\infty}e^{-x^{2}}dx, \quad\quad \lvert R(X) \rvert \leq \frac{\sqrt{\pi}}{2}e^{-X^{2}}}$.
</h5>

```python
reset()

X = 1
RX = sqrt(pi) / 2 * exp(-X^(2))
eps = 10^-3

def f(x):
    return exp(-x^(2))

def F(X):
    return numerical_integral(f, 0, X)[0]


plot(f, (0, 3), color='blue') +plot(F, (0, 3), color='red')
```

L'integrale impropre semble converger vers 0.9

```python
while RX > eps:
    X+=1
    RX = sqrt(pi) / 2 * exp(-X^(2))
    
show("Pour X=", X, ", on a I=", numerical_approx(F(X), digits=4), )
```

La réponse est cohérente avec nos observations

```python
reset()



def f(X):
    return 1 / (x - exp(-x))

def F(X):
    return numerical_integral(f(x), 1, X)[0]

plot(f(x), (1, 10), color='blue') +plot(F, (1, 10), color='red')+ plot(ln(x),(1, 10),color='green')
```

<h1>
    Exercice 3
</h1>
<h5>
    $\displaystyle{\quad\quad G(t):=\int_{0}^{+\infty}x^{t}e^{-x}dx\approx\int_{0}^{40}x^{t}e^{-x}dx, \quad\quad 0\leq t \leq 10}$.
</h5>


```python
def G(t):
    return numerical_integral(x^(t) * e^(-x), 0, 40)[0]

show(plot(G, (0, 10), ymax=10^3)) 

for t in range(0,11,1):
    print("G(",t,")=",numerical_approx(G(t), digits=4),"  ", t,"!=", factorial(t))
```

On constate que les valeurs prises par la fonction lorsque t est entier sont celle de la fonction factorielle

```python
def parties(E):

    if E ==[] : 
       return [[]]

    else : 
        x, EE = E[0], E[1:]
        
        tmp = parties(EE)
        return [ [x] + A for A in tmp] + tmp

def all_relations(n) :
    return parties([(x,y) for x in range(n) for y in range(n)])



```

```python
def mat_adj(n,rel):
    mat = matrix(n)
    for i in range(n):
        for j in range(n):
            if(i,j) in rel:
                mat[i,j] = 1
    return mat
```

```python
def all_matrices(n):
    liste = []
    for l in all_relations(n):
        liste.append( mat_adj(n,l))
    return liste
```

```python
def reflexive(n,mat):
    for i in range(n):
        if(mat[i,i] == 0):
            return 0
    return 1
    
def symetrique(mat):
    return (mat.transpose() == mat)

def antisymetrique(n, mat):
    for i in range(n):
        for j in range(n):
            if(mat[i,j]== mat[j,i] and i!= j and mat[i,j]==1):
                return 0
            
    return 1
    
def transitive(n, mat):
    for i in range(n):
        for j in range(n):
            if(mat[i,j]):
                for k in range(n):
                    if(mat[j,k]):
                        if(mat[i,k] == 0):
                            return 0
    return 1


def test_mat(n):
    a,b,c,d = 0,0,0,0
    for m in all_matrices(n):
        a += reflexive(n,m)
        b += symetrique(m)
        c += antisymetrique(n,m)
        d += transitive(n,m)
        
    print("il y a %d matrices reflexives" % a)
    print("il y a %d matrices symetrique" % b)
    print("il y a %d matrices antisymetrique" % c)
    print("il y a %d matrices transitive" % d)
```

```python
def Q(n,rand):
    A = matrix(n)
    for i in range(0,n):
        for j in range(0,n):
            if(random() <= rand):
                A[i,j] = 1
            else:
                A[i,j] = 0
    return A

def R(A,n):
    if not  reflexive(n,A):
        for i in range(n):
            A[i,i] = 1
    return A

def S(A,n):
    if not symetrique(A):
        for i in range(n):
            for j in range(n):
                if(A[i,j] and not A[j,i] ):
                    A[j,i] = 1
                        
    return A


def T(A,n):
    if not transitive(n,A):
        for i in range(n):
            for j in range(n):
                if(A[i,j]):
                    for k in range(n):
                        if(A[j,k]):
                            A[i,k]= 1
                        
    return A
```

```python
import numpy
q = Q(50,0.01)

r = R(q,50)
s = S(r,50)
t = T(s,50)


print(t)

```
